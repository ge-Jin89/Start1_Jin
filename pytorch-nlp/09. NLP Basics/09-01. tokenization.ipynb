## 2. нҶ нҒ°нҷ” мӨ‘ мғқкё°лҠ” м„ нғқмқҳ мҲңк°„

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import WordPunctTokenizer
from tensorflow.keras.preprocessing.text import text_to_word_sequence
```

```python
print('лӢЁм–ҙ нҶ нҒ°нҷ”1 :',word_tokenize("Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop."))
```

## 3. нҶ нҒ°нҷ”м—җм„ң кі л Өн•ҙм•јн•  мӮ¬н•ӯ

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()

text = "Starting a home-based restaurant may be an ideal. it doesn't have a food chain or restaurant of their own."
print('нҠёлҰ¬лұ…нҒ¬ мӣҢл“ңнҶ нҒ¬лӮҳмқҙм Җ :',tokenizer.tokenize(text))
```

## 4. л¬ёмһҘ нҶ нҒ°нҷ”(Sentence Tokenization)

```python
from nltk.tokenize import sent_tokenize

text = "His barber kept his word. But keeping such a huge secret to himself was driving him crazy. Finally, the barber went up a mountain and almost to the edge of a cliff. He dug a hole in the midst of some reeds. He looked about, to make sure no one was near."
print('л¬ёмһҘ нҶ нҒ°нҷ”1 :',sent_tokenize(text))
```

```python
text = "I am actively looking for Ph.D. students. and you are a Ph.D student."
print('л¬ёмһҘ нҶ нҒ°нҷ”2 :',sent_tokenize(text))
```

## 7. NLTKмҷҖ KoNLPyлҘј мқҙмҡ©н•ң мҳҒм–ҙ, н•ңкөӯм–ҙ нҶ нҒ°нҷ” мӢӨмҠө

```python
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

text = "I am actively looking for Ph.D. students. and you are a Ph.D. student."
tokenized_sentence = word_tokenize(text)

print('лӢЁм–ҙ нҶ нҒ°нҷ” :',tokenized_sentence)
print('н’ҲмӮ¬ нғңк№… :',pos_tag(tokenized_sentence))
```

```python
from konlpy.tag import Okt
from konlpy.tag import Kkma

okt = Okt()
kkma = Kkma()

print('OKT нҳ•нғңмҶҢ л¶„м„қ :',okt.morphs("м—ҙмӢ¬нһҲ мҪ”л”©н•ң лӢ№мӢ , м—°нңҙм—җлҠ” м—¬н–үмқ„ к°Җлҙҗмҡ”"))
print('OKT н’ҲмӮ¬ нғңк№… :',okt.pos("м—ҙмӢ¬нһҲ мҪ”л”©н•ң лӢ№мӢ , м—°нңҙм—җлҠ” м—¬н–үмқ„ к°Җлҙҗмҡ”"))
print('OKT лӘ…мӮ¬ м¶”м¶ң :',okt.nouns("м—ҙмӢ¬нһҲ мҪ”л”©н•ң лӢ№мӢ , м—°нңҙм—җлҠ” м—¬н–үмқ„ к°Җлҙҗмҡ”"))
```

```python
print('кј¬кј¬л§Ҳ нҳ•нғңмҶҢ л¶„м„қ :',kkma.morphs("м—ҙмӢ¬нһҲ мҪ”л”©н•ң лӢ№мӢ , м—°нңҙм—җлҠ” м—¬н–үмқ„ к°Җлҙҗмҡ”"))
print('кј¬кј¬л§Ҳ н’ҲмӮ¬ нғңк№… :',kkma.pos("м—ҙмӢ¬нһҲ мҪ”л”©н•ң лӢ№мӢ , м—°нңҙм—җлҠ” м—¬н–үмқ„ к°Җлҙҗмҡ”"))
print('кј¬кј¬л§Ҳ лӘ…мӮ¬ м¶”м¶ң :',kkma.nouns("м—ҙмӢ¬нһҲ мҪ”л”©н•ң лӢ№мӢ , м—°нңҙм—җлҠ” м—¬н–үмқ„ к°Җлҙҗмҡ”"))
```
